```python
cat = {1:'서울', 2:'인천', 3:'대전', 4:'대구', 5:'광주', 6:'부산', 7:'울산', 31:'경기도', 32:'강원도',
      33:'충청북도', 34:'충청남도', 35:'경상북도', 36:'경상남도', 37:'전라북도', 38:'전라남도', 39:'제주도'}
```

```python
# 시도 데이터 db에 저장
def saveDBtable(data):
    sido_num = data[1]
    sido_name = data[2]
    
    oracle_sql = '''
    INSERT INTO area_sido (sido_num, sido_name)
    VALUES (:sido_num, :sido_name)
    '''
    cursor.execute(oracle_sql, (sido_num, sido_name))
    connect.commit()


# 데이터 입력
if __name__ == '__main__':
    path = './전국지역코드.csv'
    file = open(path, 'r', encoding='utf-8')
    datas = csv.reader(file, delimiter=',')
    next(datas, None) # 한 줄을 읽은 뒤, 아무 일도 하지 않음
    for row in datas:
        print(row)
        saveDBtable(row)
```

```python
# 시군구 데이터 db에 저장

def saveDBtable2(data, sido_num):
    sigungu_num = sido_num * 100 + int(data[0])
    sigungu_name = data[2]
    print(sigungu_num)
    
    oracle_sql = '''
    INSERT INTO area_sigungu (sigungu_num, sigungu_name, area_sido_sido_num)
    VALUES (:sigungu_num, :sigungu_name, :sido_num)
    '''
    cursor.execute(oracle_sql, (sigungu_num, sigungu_name, sido_num))
    connect.commit()

for i in list(cat.keys()):
    if __name__ == '__main__':
        # 데이터 입력
        path = r'.\\시군구코드\\' + cat[i] + '시군구코드.csv'
        file = open(path, 'r', encoding='utf-8')
        datas = csv.reader(file, delimiter=',')
        next(datas, None) # 한 줄을 읽은 뒤, 아무 일도 하지 않음
        for row in datas:
            print(row)
            saveDBtable2(row, i)
```

```python
# areas 시퀀스 생성
oracle_sql = '''
    CREATE SEQUENCE area_num_seq
       INCREMENT BY 1
       START WITH 1
       MINVALUE 1
       NOCYCLE
       NOCACHE
       NOORDER
    '''
    cursor.execute(oracle_sql)
    connect.commit()
```

```python
# 위치데이터 db에 저장
def viewDBtable(sido_num, sigungu_name):
    sql='SELECT sigungu_num FROM area_sigungu WHERE area_sido_sido_num = {} and sigungu_name = \'{}\''.format(sido_num, sigungu_name)
    cursor.execute(sql)
    rows = cursor.fetchall()
    return rows
    
def saveDBtable3(data):
    area_address1 = data[0]
    area_address2 = data[1]
    area_image = data[10]
    area_image2 = data[11]
    area_mapx = data[12]
    area_mapy = data[13]
    readcount = data[16]
    area_title = data[19]
    area_overview = data[-1]
    
    sido_num = data[2]
    area_sigungu_sigungu_num = 0
    
    temp = area_address1.split(" ")
    
    if len(temp) > 1:
        sigungu_name = temp[1]
        sigungu_num_list = viewDBtable(sido_num, sigungu_name)
        if len(sigungu_num_list) > 0:
            area_sigungu_sigungu_num = list(sigungu_num_list[0])[0]
                
    #print(area_address1, area_sigungu_sigungu_num)
    
    oracle_sql = '''
    INSERT INTO AREAS (AREA_NUM, AREA_TITLE, AREA_ADDRESS1, AREA_ADDRESS2, AREA_MAPX, AREA_MAPY, AREA_IMAGE, AREA_CREATE_DATE, AREA_SIGUNGU_SIGUNGU_NUM, AREA_IMAGE2, READCOUNT, AREA_OVERVIEW)
    VALUES(area_num_seq.NEXTVAL, :area_title, :area_address1, :area_address2, :area_mapx, :area_mapy, :area_image, to_char(sysdate), :area_sigungu_sigungu_num, :area_image2, :readcount, :area_overview)
    '''
    
    if area_sigungu_sigungu_num != 0:
        cursor.execute(oracle_sql, (area_title, area_address1, area_address2, area_mapx, area_mapy, area_image, area_sigungu_sigungu_num, area_image2, readcount, area_overview))
        connect.commit()

for sido_num in list(cat.keys()):
    # 폴더별 파일 이름 읽기
    if __name__ == '__main__':
        res = []
        dir_path = ".\\지역별관광정보데이터\\" + cat[sido_num] + "\\시군구지역_관광정보(내용포함)"
        for path in os.listdir(dir_path):
            if os.path.isfile(os.path.join(dir_path, path)):
                res.append(path)
        #print(res)
        
        # 데이터 입력
        for j in res:
            try:
                path = r'.\\지역별관광정보데이터\\' + cat[sido_num] + '\\시군구지역_관광정보(내용포함)\\' + j
                file = open(path, 'r', encoding='utf-8')
                datas = csv.reader(file, delimiter=',')
                next(datas, None) # 한 줄을 읽은 뒤, 아무 일도 하지 않음
                for row in datas:
                    if row[10] != '':
                        #print(row)
                        saveDBtable3(row)
            except UnicodeDecodeError as e:
                print(e)
                continue
```
